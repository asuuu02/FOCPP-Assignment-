```python
import random
```

```python
def check_password_length(password):
    if len(password) < 9:
        print("Password too short.")
        return False
    return True
```

```python
def security_check(password):
    for _ in range(3):
        position = random.randint(1, len(password))
        letter = input(f"Enter letter at position {position}: ")
        
        if letter == password[position - 1]:
            print("Correct")
        else:
            print("Security check failed.")
            return False
    
    return True
```

```python
password = input("Enter your password: ")

if check_password_length(password):
    if security_check(password):
        print("Security check passed.")
```

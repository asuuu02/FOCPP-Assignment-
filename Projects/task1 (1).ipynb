```python
def get_valid_price(pizza_number):
    while True:
        try:
            price = float(input(f"Enter Price of Pizza #{pizza_number}: "))
            if price > 0:
                return price
            else:
                print("Please enter a valid price!")
        except ValueError:
            print("Please enter a valid price!")
```

```python
pizza_prices = []

for i in range(1, 5):
    pizza_prices.append(get_valid_price(i))
```

```python
original_total = sum(pizza_prices)
cheapest_pizza = min(pizza_prices)

final_total = original_total - cheapest_pizza
discount_percentage = round((cheapest_pizza / original_total) * 100)
```

```python
width = 55

print("\n" + "Beckett Pizza Plaza 4-for-3 Offer".center(width))
print("=================================".center(width))
print()

for i, price in enumerate(pizza_prices, start=1):
    print(f"Enter Price of Pizza #{i}: {price:.2f}".center(width))

print()
print(f"Order Total is £{final_total:.2f}, a fabulous discount of {discount_percentage}%!".center(width))
```

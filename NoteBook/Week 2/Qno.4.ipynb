```python
total_sweets = int(input("How many sweets are in the tub? "))
num_pupils = int(input("How many pupils are present today? "))
```

```python
sweets_each = total_sweets // num_pupils
leftover = total_sweets % num_pupils
```

```python
print("\nSweet Distribution Calculator")
print("-----------------------------")
print(f"Each pupil gets: {sweets_each} sweets")
print(f"Sweets left over: {leftover}\n")
```

```python
import statistics  

temperatures = []

for i in range(6):
    user_input = input(f"Enter temperature {i+1} in Celsius (e.g., 25C): ")
    temp = float(user_input.rstrip('Cc'))
    temperatures.append(temp)

maximum_temp = max(temperatures)
minimum_temp = min(temperatures)
mean_temp = statistics.mean(temperatures)

print(f"Maximum temperature: {maximum_temp}C")
print(f"Minimum temperature: {minimum_temp}C")
print(f"Mean temperature: {mean_temp:.2f}C")
```

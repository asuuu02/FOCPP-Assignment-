```python
def unique_letters_sorted(s):

    letters_only = [char.lower() for char in s if char.isalpha()]
    
    unique_sorted = sorted(set(letters_only))
    
    return unique_sorted
```

```python
test_strings = ["cheese", "Python Programming!", "Cat"]

for string in test_strings:
    print(f"Original: '{string}' -> Unique sorted letters: {unique_letters_sorted(string)}")
```

```python
user_input = input("Enter a string: ")
print("Unique sorted letters:", unique_letters_sorted(user_input))
```

```python
def letters(word):
    return set(char.lower() for char in word if char.isalpha())
```

```python
def letters_union(word1, word2):
    return sorted(letters(word1) | letters(word2)) 
```

```python
def letters_intersection(word1, word2):
    return sorted(letters(word1) & letters(word2))
```

```python
def letters_symmetric_difference(word1, word2):
    return sorted(letters(word1) ^ letters(word2)) 
```

```python
word1 = input("Enter the first word: ")
word2 = input("Enter the second word: ")

print("\nLetters in at least one of the words (Union):", letters_union(word1, word2))
print("Letters in both words (Intersection):", letters_intersection(word1, word2))
print("Letters in either word but not both (Symmetric Difference):", letters_symmetric_difference(word1, word2))
```

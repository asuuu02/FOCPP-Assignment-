```python
def factors(n):

    if n == 0:
        return [] 
    n = abs(n) 
    result = []
    for i in range(1, n+1):
        if n % i == 0:
            result.append(i)
    return result
```

```python
test_numbers = [6, 12, 15, -10, 0]

for num in test_numbers:
    print(f"Factors of {num}: {factors(num)}")
```

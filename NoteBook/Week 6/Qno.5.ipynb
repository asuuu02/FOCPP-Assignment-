```python
import random
import string
```

```python
def hidden_message_encrypt(message):
    
    if not message:
        return "", 0

    interval = random.randint(2, 20) 
    encrypted = ""

    for char in message:
        encrypted += char  
        for _ in range(interval - 1):
            encrypted += random.choice(string.ascii_lowercase)
    
    return encrypted, interval
```

```python
message = "send cheese"
encrypted_message, interval_used = hidden_message_encrypt(message)

print(f"Original message: {message}")
print(f"Encrypted message: {encrypted_message}")
print(f"Interval used: {interval_used}")
```

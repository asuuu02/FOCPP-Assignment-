```python
import sys
import shutil
import os
```

```python
sys.argv = ['Qno.6.ipynb', 'Qno.5.ipynb']
```

```python
if len(sys.argv) < 2:
    print("Please provide the name of a file as a command-line argument.")
    sys.exit(1)

original_file = sys.argv[1]

name, ext = os.path.splitext(original_file)
backup_file = f"{name}_backup{ext}"
```

```python
try:
    shutil.copy(original_file, backup_file)
    print(f"Backup created: {backup_file}")
except FileNotFoundError:
    print(f"The file '{original_file}' does not exist!")
except Exception as e:
    print(f"An error occurred: {e}")
```

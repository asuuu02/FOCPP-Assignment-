```python
import sys
```

```python
sys.argv = ["Qno.2.ipynb", "File.1.txt", "File.2.txt"]
```

```python
if len(sys.argv) < 3:
    print("Error: Two file names are required.")
else:
    file1 = sys.argv[1]
    file2 = sys.argv[2]

    try:
        with open(file1, "r") as f1, open(file2, "r") as f2:
            content1 = f1.read()
            content2 = f2.read()

            if content1 == content2:
                print("Files are the same.")
            else:
                print("Files are different.")

    except FileNotFoundError:
        print("Error: One or both files not found.")
```

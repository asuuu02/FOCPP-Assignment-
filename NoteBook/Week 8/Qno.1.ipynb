```python
import sys
```

```python
sys.argv = ["Qno.1.ipynb", "File.1.txt"]
```

```python
if len(sys.argv) < 2:
    print("Error: No file name provided.")
else:
    filename = sys.argv[1]

    try:
        with open(filename, "r") as file:
            line_number = 1
            for line in file:
                print(f"{line_number}\t{line.rstrip()}")
                line_number += 1
    except FileNotFoundError:
        print("Error: File not found.")
```

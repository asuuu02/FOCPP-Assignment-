```python
import sys
```

```python
sys.argv = ["Qno.3.ipynb", "Python", "File.3.txt"]
```

```python
if len(sys.argv) < 3:
    print("Error: Search string and file name required.")
else:
    pattern = sys.argv[1]
    filename = sys.argv[2]

    try:
        with open(filename, "r") as file:
            for line in file:
                if pattern.lower() in line.lower():
                    print(line.rstrip())
    except FileNotFoundError:
        print("Error: File not found.")
```
